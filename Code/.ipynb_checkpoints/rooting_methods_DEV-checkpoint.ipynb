```python
def mp_from_clades(clade1, clade2, max_distance, tree):
    # Depth to go from the ingroup tip toward the outgroup tip 
    root_remainder = 0.5 * (max_distance - (tree.root.branch_length or 0)) 
    assert root_remainder >= 0 
    ####This is straight from the biopython midpoint root algorithm
    # Identify the midpoint and reroot there. 
    # Trace the path to the outgroup tip until all of the root depth has 
    # been traveled/accounted for. 
    for node in tree.get_path(clade2): 
        root_remainder -= node.branch_length 
        if root_remainder < 0: 
            outgroup_node = node 
            outgroup_branch_length = -root_remainder
            break 
    else: 
        raise ValueError("Somehow, failed to find the midpoint!") 
    tree.root.clades[0].branch_length = tree.root.clades[0].branch_length-outgroup_branch_length
    tree.root_with_outgroup(outgroup_node, outgroup_branch_length=outgroup_branch_length)
    return tree
```

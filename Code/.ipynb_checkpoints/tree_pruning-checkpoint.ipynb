```python
n_to_prune = 15
for infile in glob.glob('../Data/euk_trees/*.nwk.Rooted.MPAJH')[:]:
    print(infile)
    ###Using ete3
    tree = ete3.Tree(infile)
#     to_remove = random.sample(metazoa, n_to_prune) 
    to_remove = random.sample(not_metazoa, n_to_prune) 
    for i in to_remove:
        G = tree.search_nodes(name=i)[0]
        G.delete()
    outfile_loc = infile.replace('euk_trees/', 'pruned_euk_trees/')
    outfile_loc = outfile_loc.replace('.nwk.Rooted.MPAJH', '.pruned_{}_notmeta.nwk'.format(n_to_prune))
    print(outfile_loc)
    tree.write(format=1, outfile=outfile_loc)
```
